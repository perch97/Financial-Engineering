```python
from statsmodels.graphics.tsaplots import plot_pacf,plot_acf
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x0 = np.random.randn(1000)
```

```python
plt.plot(x0)
```

```python
x0.shape
```

```python
fig, ax = plt.subplots(figsize=(10,5))
plot_pacf(x0,ax=ax);
```

```python
x1 = [0]
for i in range(1000):
    x = 0.5*x1[-1]+0.1*np.random.randn()
    x1.append(x)
x1=np.array(x1)
```

```python
plt.plot(x1)
```

```python
fig,ax=plt.subplots(figsize=(10,5))
plot_pacf(x1,ax=ax);
```

```python
x1=[0]
for i in range(1000):
    x = -0.5*x1[-1]+0.1*np.random.randn()
    x1.append(x)
x1 = np.array(x1)
```

```python
plt.plot(x1);
```

```python
fig,ax=plt.subplots(figsize=(10,5))
plot_pacf(x1,ax=ax);
```

```python
x2=[0,0]
for i in range(1000):
    x = 0.5*x2[-1]-0.3*x2[-2]+0.1*np.random.randn()
    x2.append(x)
x2=np.array(x2)
```

```python
#iid noise
fig,ax = plt.subplots(figsize=(10,5))
plot_acf(np.random.randn(1000),ax=ax);
```

```python
errors = 0.1*np.random.randn(1000)
ma1=[]
for i in range(1000):
    if i>=1:
        x=0.5*errors[i-1]+errors[i]
    else:
        x=errors[i]
    ma1.append(x)
ma1=np.array(ma1)
```

```python
plt.plot(ma1);
```

```python
fig,ax = plt.subplots(figsize=(10,5))
plot_acf(ma1,ax=ax);
```
